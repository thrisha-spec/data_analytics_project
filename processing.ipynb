```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
sns.countplot(x='Gender',data=df,color="red")
plt.title("Gender")
plt.show
```

```python

plt.figure(figsize=(10,4))
sns.barplot(x="Gender",y='Age',data=df,color='pink')
plt.title('mental health and social media')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```
